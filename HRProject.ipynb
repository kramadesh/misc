```python
#Staff class to keep track of all the staff memebers, including the manager.
"""
staff class has follwing attributes:
firstname,
lastname,
date of birth,
zipcode,
hiring date
"""
class Staff:
    def __init__(self, fname, lname, phone, email, dob, zipcode, hiringdate) -> None:
        self.fname = fname
        self.lname = lname
        self.name = fname +" " +lname
        self.dob = dt.strptime(dob,'%m/%d/%Y').date()
        self.phone = phone
        self.email = email
        self.zipcode = zipcode
        self.hiringdate = dt.strptime(hiringdate,'%m/%d/%Y').date()

    def display(self):
        #print(f'Name {" "*10} Lname{" "*10} DoB Zipcode{" "*6} Hiringdate')
        #print(f"{self.name:20} {self.fname:10} {self.lname:10} {self.dob} {self.zipcode:6} {self.hiringdate}")
        detail = f"{self.name:20} {self.fname:10} {self.lname:10} {self.dob} {self.zipcode:6} {self.hiringdate}"
        return detail
        
   
```

```python
#Manager class. In inherits from the Staff class
# It also has subordinate staff. Subordinate staffs can be added and removed
# show

class Manager(Staff):
    def __init__(self, fname, lname, phone, email, dob, zipcode, hiringdate) -> None:
        super().__init__(fname, lname, phone, email, dob, zipcode, hiringdate)
        self.manager = True
        self.staff = set()
        self.name = fname +" " +lname

    def addSubordinate(self, member):
        self.staff.add(member)
    
    def removeSubordinate(self, member):
        to_remove = None
        for s in self.staff:
            if s.fname ==member.fname and s.lname == member.lname:
                to_remove = s   
        if to_remove:
            self.staff.remove(to_remove)
        else:
            print("No matching subordinate found")

    def showSubOrdinates(self):
        print(f"Showing all the subordinates for this Manager {self.name}")
        for s in self.staff:
            print(s.display())
```
