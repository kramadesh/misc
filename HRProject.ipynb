```python
#Staff class to keep track of all the staff memebers, including the manager.
"""
staff class has follwing attributes:
firstname,
lastname,
date of birth,
zipcode,
hiring date
"""
class Staff:
    def __init__(self, fname, lname, phone, email, dob, zipcode, hiringdate) -> None:
        self.fname = fname
        self.lname = lname
        self.name = fname +" " +lname
        self.dob = dt.strptime(dob,'%m/%d/%Y').date()
        self.phone = phone
        self.email = email
        self.zipcode = zipcode
        self.hiringdate = dt.strptime(hiringdate,'%m/%d/%Y').date()

    def display(self):
        print(f"{self.name:20} {self.fname:10} {self.lname:10} {self.dob} {self.zipcode:6} {self.hiringdate}")
        
   
```

```python
#Manager class. In inherits from the Staff class
# It also has subordinate staff. Subordinate staffs can be added and removed
# show

class Manager(Staff):
    def __init__(self, fname, lname, phone, email, dob, zipcode, hiringdate) -> None:
        super().__init__(fname, lname, phone, email, dob, zipcode, hiringdate)
        self.manager = True
        self.staff = set()
        self.name = fname +" " +lname

    def addSubordinate(self, member):
        self.staff.add(member)
    
    def removeSubordinate(self, member):
        to_remove = None
        for s in self.staff:
            if s.fname ==member.fname and s.lname == member.lname:
                to_remove = s   
        if to_remove:
            self.staff.remove(to_remove)
        else:
            print("No matching subordinate found")

    def showSubOrdinates(self):
        print(f"Showing all the subordinates for this Manager {self.name}")
        for s in self.staff:
            s.display()
```

```python
#Implementation of system directory.
"""
SystemDirectory maintains a directory of staff members.
Staff can be added and removed.
All the members of the directory can be displayed at once.
Members 
"""
class SystemDirectory:
    def __init__(self, hradmin) -> None:
        self.HRAdmin = hradmin
        self.staffDir = [hradmin]

    #Functon to add a staff to the System directory
    def addStaff(self, staff):
        index = 0
        for s in self.staffDir:
            if s.fname  == staff.fname and s.lname == staff.lname and s.dob == staff.dob:
                print(f"Cannot add staff {staff.name}. Entry already exists. ")
                return
            else:
                if s.hiringdate < staff.hiringdate:
                    index+=1
        self.staffDir.insert(index, staff)
        print(f"Staff {staff.name} added to the Staff Directory")
    #Functon to remove a staff fromt he directory 
    def removeStaff(self, S):
        removed=False
        for s in self.staffDir:
            if s.fname ==S.fname and s.lname == S.lname:
                self.staffDir.remove(s)
                print(f"Staff {s.name} removed from the Staff Directory")
                removed = True
        if removed==False:
            print("No matching staff found")
    #Function to display the System Directory
    def showSystemDirectory(self):
        print("############Printing the complete directory############")
        for s in self.staffDir:
            s.display()
        print("############Directory printed.############")
    #Function to search staff by Name
    def searchStaffByName(self, staff):
        stafflist=[]
        for s in self.staffDir:
            if s.fname ==staff.fname and s.lname == staff.lname:
                stafflist.append(s)
        if len(stafflist)==0:
            print("No match found.")
            return
        else:
            for p in stafflist:
                p.display()
    #Function to search staff by Hiring Date
    def searchStaffByHiringDate(self, hdate):
        stafflist=[]
        hdate = dt.strptime(hdate,'%m/%d/%Y').date()
        for s in self.staffDir:
            if s.hiringdate ==hdate:
                stafflist.append(s)
        if len(stafflist)==0:
            print("No match found.")
            return
        else:
            for p in stafflist:
                p.display()
    #Function to search staff by birthday
    def searchStaffByBirthdate(self, dob):
        stafflist=[]
        hdate = dt.strptime(dob,'%m/%d/%Y').date()
        for s in self.staffDir:
            if s.dob ==dob:
                stafflist.append(s)
        if len(stafflist)==0:
            print("No match found.")
            return
        else:
            for p in stafflist:
                p.display()
    #Function to search staff by Zipcode
    def searchStaffByZipcode(self, zipcode):
        stafflist=[]
        for s in self.staffDir:
            if s.zipcode == zipcode:
                stafflist.append(s)
        if len(stafflist)==0:
            print("No match found.")
            return
        else:
            for p in stafflist:
                p.display()
```
